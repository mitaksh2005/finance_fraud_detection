```python
# Cell 3
import pandas as pd
import numpy as np

def reduce_mem_usage(df):
    for col in df.columns:
        col_type = df[col].dtype
        if col_type != object:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
            else:
                df[col] = df[col].astype(np.float32) # Standard for TabNet
    return df
```

```python
# Cell 4
# Load only Training data first to save RAM
train_transaction = pd.read_csv('../data/raw/train_transaction.csv')
train_identity = pd.read_csv('../data/raw/train_identity.csv')

# Merge on TransactionID
train = pd.merge(train_transaction, train_identity, on='TransactionID', how='left')

# Apply memory reduction
train = reduce_mem_usage(train)
print(f"Memory usage after downcasting: {train.memory_usage().sum() / 1024**2:.2f} MB")
```

```python
# Cell 5
# Load only Training data first to save RAM
test_transaction = pd.read_csv('../data/raw/test_transaction.csv')
test_identity = pd.read_csv('../data/raw/test_identity.csv')

# Merge on TransactionID
test = pd.merge(test_transaction, test_identity, on='TransactionID', how='left')

# Apply memory reduction
test = reduce_mem_usage(test)
print(f"Memory usage after downcasting: {test.memory_usage().sum() / 1024**2:.2f} MB")
```

```python
#Cell 6
test.rename(columns=lambda x: x.replace('-', '_'), inplace=True)

train_cols = set(train.columns)
test_cols = set(test.columns)
common_cols = train_cols.intersection(test_cols)

expected_diff = {'isFraud'}
actual_diff = train_cols - test_cols

assert actual_diff == expected_diff, f"Column mismatch! Check: {actual_diff - expected_diff}"
print("Columns are perfectly aligned.")
```

```python
# Cell 7
# Save as Parquet for Phase 2
train.to_parquet('../data/processed/train.parquet')
test.to_parquet('../data/processed/test.parquet')
```
